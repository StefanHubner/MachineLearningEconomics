```r
# Load necessary libraries
library(ggplot2)

# Set sample size
N <- 200

# 1. Simulate bivariate data x1 and x2 with uniform distribution [0,1]
set.seed(123)  # for reproducibility
x1 <- runif(N, 0, 1)
x2 <- runif(N, 0, 1)

# 2. Construct ystar - a linear function of x1 and x2 with a normal error term
# We'll use coefficients 0.5 for x1 and x2 for demonstration purposes. Adjust as needed.
epsilon <- rnorm(N, 0, 0.1)  # normal error term with mean 0 and sd 0.1
ystar <- 0.5 * x1 + 0.5 * x2 + epsilon

# 3. Classify y = 0 if ystar > 1/2, otherwise 1
y <- ifelse(ystar > 0.5, 0, 1)

# 4. Plot with ggplot2
data <- data.frame(x1, x2, y)
p <- ggplot(data, aes(x=x1, y=x2, color=factor(y))) +
  geom_point() +
  scale_color_discrete(name="Class") +
  labs(title="Scatter plot of x1 vs x2 colored by class", x="x1", y="x2") +
  geom_abline(intercept = 1, slope = -1, linetype="dashed", color="black", linewidth=1)

print(p)
```

Given:
$$ y^* = 0.5 x_1 + 0.5 x_2 + \epsilon $$
where $ \epsilon \sim N(0, 0.1^2) $,

The Bayes decision boundary is derived where $ y^* = 0.5 $. So, setting them equal:
$$ 0.5 x_1 + 0.5 x_2 + \epsilon = 0.5 $$

Given that $ \epsilon $ is a random noise term with an expected value (mean) of 0, this can be simplified further to:
$$ 0.5 x_1 + 0.5 x_2 = 0.5 $$

Now, isolating $ x_2 $:
$$ x_2 = 1 - x_1 $$

And that's the equation for the Bayes decision boundary in the given $ x_1, x_2$.


```r
library(ggplot2)

# Set sample size
N <- 200

# Simulate data and generate y using the quadratic specification
set.seed(123)
x1 <- runif(N, 1, 2)
x2 <- runif(N, -1, 0)
epsilon <- rnorm(N, 0, 0.1)
ystar <- x1 + 2 * x2 + x2^2 + epsilon
y <- ifelse(ystar > 0.5, 0, 1)

# Plot data with ggplot2
data <- data.frame(x1, x2, y)
p <- ggplot(data, aes(x=x1, y=x2, color=factor(y))) +
  geom_point() +
  geom_line(aes(x=x1, y=-1 + sqrt(3/2 - x1)), color="black", linetype="dashed", size=1) +
  scale_color_discrete(name="Class") +
  labs(title="Scatter plot of x1 vs x2 (quadratic) with Bayes Decision Boundary", x="x1", y="x2")

suppressWarnings({print(p)})
```

Given:
$$ y^* = x_1 + 2 x_2 + x_2^2 + \epsilon $$
where $ \epsilon \sim N(0, 0.1^2) $,

The Bayes decision boundary is derived where $ y^* = 0.5 $. So, setting them equal:
$$ x_1 + 2 x_2 + x_2^2 + \epsilon = 0.5 $$

Given that $ \epsilon $ is a random noise term with an expected value (mean) of 0, this can be simplified further to:
$$  x_2^2  + 2 x_2 + (x_1 - 0.5) = 0 $$

Now, solving for $ x_2 $:
$$ x_2 = -\frac{p}{2} \pm \sqrt{\frac{p^2}{4}-q} =  - 1 \pm \sqrt{ \frac{3}{2} - x_1} $$

And that's the equation for the Bayes decision boundary in the given $ x_1, x_2$.
